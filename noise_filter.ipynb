```python
#read img
img = cv2.imread('images.jpg')

#convert img to grayscale
gray_img = img[:,:,1]
#generate noise with same size of that image
noise = np.random.normal(0,50,gray_img.shape)
#add noise to image
img_noised = gray_img + noise
#clip the pixel value in between 0 to 255
clip = np.clip(img_noised,0,255).astype(np.uint8)
plt.imshow(img_noised,cmap='gray')
```

```python
#salt and peper
#get the image size(num of pixel in image)
img_size = gray_img.size
#get the noise percentage
noise_percentage = 0.1
# Determine the size of the noise based on the noise precentage
noise_size = int(noise_percentage * img_size)
#randomly select indces for adding noise
random_indices = np.random.choice(img_size, noise_size)
# Create a copy of the original image that serves as a template for the noised image.
img_noised = gray_img.copy()
# Create a noise list with random placements of min and max values of the image pixels.
noise = np.random.choice([gray_img.min(), gray_img.max(), noise_size])
# Replace the values of the templated noised image at random indices with the noise, to obtain the final noised image.
img_noised.flat[random_indices] = noise
plt.imshow(img_noised,cmap='gray')
```

```python
#denoise the salt and peper filter
#get the image size(num of pixel in image)
img_size = gray_img.size
#get the noise percentage #in order to denoise image we set the noise percentage to 0
noise_percentage = 0
# Determine the size of the noise based on the noise precentage
noise_size = int(noise_percentage * img_size)
#randomly select indces for adding noise
random_indices = np.random.choice(img_size, noise_size)
# Create a copy of the original image that serves as a template for the noised image.
img_noised = gray_img.copy()
# Create a noise list with random placements of min and max values of the image pixels.
noise = np.random.choice([gray_img.min(), gray_img.max(), noise_size])
# Replace the values of the templated noised image at random indices with the noise, to obtain the final noised image.
img_noised.flat[random_indices] = noise
plt.imshow(img_noised,cmap='gray')
```
